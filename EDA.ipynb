# Exploratory Data Analysis (EDA) of Stock Market Data

This notebook guides you through the process of exploring the stock market data for insightful trends, patterns, and potential anomalies.

Sections:
1. Loading the Data
2. Overview of the Data
3. Trends in Stock Prices
4. Volume Analysis
5. Moving Averages
6. Correlations and Relationships

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ensure that plots are displayed inline in the Jupyter notebook
%matplotlib inline

# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

```python
# 1. Loading the Data
# -------------------
# Load the preprocessed stock data from a CSV file
data = pd.read_csv("aapl_preprocessed.csv")

# 2. Overview of the Data
# -----------------------
# Display the first few rows of the dataset
print(data.head())

# Get a summary of the dataset
data.describe()
```

```python
# 3. Trends in Stock Prices
# -------------------------
# Plot the closing price of the stock over time
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.title('Closing Price of AAPL Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.show()
```

```python
# 4. Volume Analysis
# ------------------
# Plot the trading volume over time
plt.figure(figsize=(14, 7))
plt.bar(data['Date'], data['Volume'], color='gray')
plt.title('Trading Volume of AAPL')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
# 5. Moving Averages
# ------------------
# Plot closing price along with moving averages
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.plot(data['Date'], data['moving_avg_5'], label='5-Day Moving Average', linestyle='--')
plt.plot(data['Date'], data['moving_avg_20'], label='20-Day Moving Average', linestyle='--')
plt.title('Closing Price and Moving Averages of AAPL')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
# 6. Correlations and Relationships
# ---------------------------------
# Analyze the correlation between different features of the data
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Stock Data Features')
plt.show()
```
